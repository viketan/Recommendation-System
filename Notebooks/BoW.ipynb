```python
import gensim
import pandas as pd
import nltk 
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from sklearn.feature_extraction.text import CountVectorizer

%matplotlib inline
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
```

```python
course_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-ML321EN-SkillsNetwork/labs/datasets/course_processed.csv"
course_content_df = pd.read_csv(course_url)
```

```python
course_content_df.head()
```

```python
course_content_df['course text'] = course_content_df[['TITLE','DESCRIPTION']].agg(' '.join, axis=1)
```

```python
course_content_df = course_content_df.reset_index()
```

```python
course_content_df
```

```python
stopwords = set(stopwords.words('english'))
porter = PorterStemmer()
```

```python
def preprocess_text(text):
    text = text.lower()
    tokens = word_tokenize(text)
    filtered_tokens = [porter.stem(token) for token in tokens if token.isalnum() and token not in stopwords]
    return ' '.join(filtered_tokens)
```

```python
course_content_df['Processed_text'] = course_content_df['course text'].apply(preprocess_text)
```

```python
course_content_df
```

```python
vectorizer = CountVectorizer()
```

```python
X = vectorizer.fit_transform(course_content_df['Processed_text'])
```

```python
bow_df = pd.DataFrame(X.toarray(),columns = vectorizer.get_feature_names_out(), index= course_content_df['COURSE_ID'])
```
